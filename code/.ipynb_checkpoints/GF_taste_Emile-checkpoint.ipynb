# Are gluten-free beers worse than traditional beers?

One of the stereotypes that can be found on gluten-free beers is that they taste different than normal beers, and most often taste worse. To take a look at this matter, we can look at the difference in overall score, palate, aroma and taste scores between the gluten-free and normal beers. However, all beer types are not equally translated into gluten-free beers and it is very possible that this as a consequence on the user's ratings. To see that, we can compare the gluten-free datasets with a subset of the all-beers that has the same composition of beer type than the gluten-free one.

### Creation of BA subdataset

```python
# Create dataframes for to compare for each dataset
# BA
n_BA= 150000 #Define how many ratings we want for the subset of BA
BA_adjstyle_n = round(BA_gf_style_proportion * n_BA)
BA_adjstyle_n.sum() #=100 001, considered OK
BA_adjstyle_n #Number of ratings from each beer style we want
```

```python
BA_style_n=BA_ratings['style'].value_counts()
BA_style_n
```

```python
#Look at whether there is enough beers in BA_ratings to create the subset 
for style in BA_gf_style_proportion.index:
    if BA_style_n[style] >= BA_adjstyle_n[style]:
        print(style,' : OK')
    else:
        print(style,' : NO ________')
```

There is enough ratings in the dataset to get the adjusted subset we want (n= 150 000)

```python
BA_adjtype_ID =[] #List of ID of the selected comments
for style in BA_gf_style_proportion.index:
    BA_1style = BA_ratings_style.index[BA_ratings_style['style']==style].tolist() #Get the index of all comments from a style of beer
    BA_selectratings1style= random.sample(BA_1style,int(BA_adjstyle_n[style])) 
    for id in BA_selectratings1style:
        BA_adjtype_ID.append(id)

BA_ratings_adjstyle = BA_ratings_style.loc[BA_adjtype_ID]
```

```python
# See whether the data are distributed as we want
test_count=BA_ratings_adjstyle['style'].value_counts()
for style in test_count.index:
    if test_count[style] == BA_adjstyle_n[style]:
        print(style,' : OK')
    else:
        print(style,' : NO ________')
```

```python
#Create year column for all desired datasets
dtset= [BA_ratings_adjstyle, RB_ratings_adjstyle]
for dt in dtset:
    dt['year']=dt['date'].apply(lambda x: int(x[:4]))
dtset = [BA_ratings_adjstyle, RB_ratings_adjstyle,BA_gf_ratings, RB_gf_ratings]
for dt in dtset:
    for col in ['rating','appearance','aroma','palate','taste','overall']:
        dt[col]=dt[col].astype('float',copy=False)
```

```python
def calculate_numerical_stats_by_year(df):
    '''
    df: filtered dataframe with numerical columns
    '''
    # Define the aggregation functions
    agg_funcs = {
        'rating': ['mean', 'std', 'sem'],
        'appearance': ['mean', 'std', 'sem'],
        'aroma': ['mean', 'std', 'sem'],
        'palate': ['mean', 'std', 'sem'],
        'taste': ['mean', 'std', 'sem'],
        'overall': ['mean', 'std', 'sem']
    }

    # Perform aggregation
    stats = df.groupby('year').agg(agg_funcs)

    # Rename the columns for clarity (optional)
    stats.columns = [f'{col[0]}_{col[1]}' if isinstance(col, tuple) else col for col in stats.columns]

    # Reset index if needed
    stats.reset_index(inplace=True)

    return stats
```

```python
#Have to create a "year" column
#metrics=['year','rating','appearance','aroma','palate','taste','overall']
RB_gf_stats = calculate_numerical_stats_by_year(RB_gf_ratings)
RB_ngf_stats = calculate_numerical_stats_by_year(RB_ratings_adjstyle)
BA_gf_stats = calculate_numerical_stats_by_year(BA_gf_ratings)
BA_ngf_stats = calculate_numerical_stats_by_year(BA_ratings_adjstyle)
```

```python
# plt the mean numerical stats over time for glutenfree and non-glutenfree beers
# plot rating, appearance, aroma, palate, taste, overall in 6 subplots
# plot the mean and the standard error of the mean (sem)

def plot_statistics(ax, x1, x2, y1, y2, yerr1, yerr2, label1, label2, xlabel, ylabel, title):
    ax.fill_between(x1, y1-yerr1, y1+yerr1, alpha=0.3)
    ax.fill_between(x2, y2-yerr2, y2+yerr2, alpha=0.3)
    ax.plot(x1,y1,label=label1)
    ax.plot(x2,y2, label=label2)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    ax.legend()

# Assuming gf_stats and ngf_stats are your DataFrames

fig, axs = plt.subplots(2, 3, figsize=(15, 10))

statistics = ['rating', 'appearance', 'aroma', 'palate', 'taste', 'overall']

for i, stat in enumerate(statistics, start=1):
    row = (i - 1) // 3
    col = (i - 1) % 3

    plot_statistics(
        axs[row, col],
        RB_gf_stats['year'],
        RB_ngf_stats['year'],
        RB_gf_stats[f'{stat}_mean'],
        RB_ngf_stats[f'{stat}_mean'],
        RB_gf_stats[f'{stat}_sem'],
        RB_ngf_stats[f'{stat}_sem'],
        'glutenfree',
        'non-glutenfree',
        'Year',
        stat.capitalize(),
        f'{stat.capitalize()} over time'
    )
    
    step = 5
    axs[row, col].set_xticks(RB_ngf_stats['year'].astype(int).unique()[::step])

plt.suptitle('Metrics over time for RateBeer dataset')
plt.tight_layout()
plt.savefig('data/MetricsRB.png')
plt.show()
```
