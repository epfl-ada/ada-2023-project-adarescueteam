```python
lines_to_keep=[]
for i in range(0,17):
    new=[position*17+i for position in positions]
    lines_to_keep=lines_to_keep+new
lines_to_keep.sort()
lines_to_keep[1:20]
```

```python
input_file_path = 'ratings.txt'
output_file_path = 'micro_breweries.txt'

import linecache
with  open(output_file_path, 'w') as output_file:
    for line_number in lines_to_keep:
        line = linecache.getline(input_file_path, line_number)
        output_file.write(line)
# Close the output file
output_file.close()
```
